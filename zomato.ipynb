#### Removing "NEW" , "-" and "/5" from Rate Column

```python
def ratehandle(value):
    if (value == "NEW" or value == "-"):
            return np.nan
    else:
        value = str(value).split("/")
        value = value[0]
        return float(value)
    
df['rate'] = df["rate"].apply(ratehandle)
```


The code defines a function ratehandle that takes a value as input and processes it. If the value is "NEW" or "-", it returns NaN. Otherwise, it splits the value by "/", takes the first part, converts it to a float, and returns the result. The apply method is then used to apply this function to the "rate" column of a DataFrame

 #### we are cleaning rest-type column to analyze further

```python
rest_type = df["rest_type"].value_counts(ascending = False)
```

```python
rest_type_less_than1000 = rest_type[rest_type<1000]
rest_type_less_than1000
```

#### Making Rest Types less than 1000 in frequency as others

```python
def handling_rest_type(value):
    if (value in rest_type_less_than1000):
        return "Others"
    else:
        return value
df["rest_type"] = df["rest_type"].apply(handling_rest_type)
```

```python
#here we can see others and ist make to easy to analyase

df['rest_type'].value_counts()
```

### Cleaning cuisines Column

```python
cuisines = df["cuisines"].value_counts()
cuisines_under100 = cuisines[cuisines <100]

def handling_cuisines(value):
    if (value in cuisines_under100 ):
        return "Others"
    else:
        return value
df["cuisines"] = df["cuisines"].apply(handling_cuisines)
```

### Cleaning cuisines Column

```python
location = df.location.value_counts()
location_under300 = location[location<300]

def handling_location(value):
    if(value in location_under300):
        return "Others"
    else:
        return value
df["location"] = df["location"].apply(handling_location)
```
